```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")

penguins = penguins.dropna()

# Display first few rows
penguins.head()
```

```python
sns.set(style="ticks")
sns.pairplot(
    penguins,
    hue="species",
    palette="bright",
    diag_kind="kde",
    markers=["o", "s", "D"]
)

plt.suptitle("Pairplot — Penguins Dataset", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    palette="deep",
    s=70
)
plt.title("Bill Length vs Bill Depth")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="flipper_length_mm",
    hue="species",
    palette="deep",
    s=70
)
plt.title("Bill Length vs Flipper Length")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species",
    palette="deep",
    s=70
)
plt.title("Flipper Length vs Body Mass")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="species", data=penguins, palette="pastel")
plt.title("Number of Penguins by Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
features = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]

plt.figure(figsize=(14,10))
for i, f in enumerate(features, 1):
    plt.subplot(2,2,i)
    sns.boxplot(x="species", y=f, data=penguins)
    plt.title(f"{f} by Species")

plt.tight_layout()
plt.show()
```

```python
penguins.hist(figsize=(12,8), bins=15, edgecolor="black")
plt.suptitle("Feature Histograms — Penguins", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    penguins.select_dtypes("number").corr(),
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)
plt.title("Correlation between Numeric Features")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.catplot(
    x="species",
    y="body_mass_g",
    hue="sex",
    kind="swarm",
    data=penguins,
    palette="Set2"
)
plt.title("Body Mass by Species & Sex")
plt.show()
```
